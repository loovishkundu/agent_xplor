## Environment Setup & API Quickstart

```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()

model = "claude-sonnet-4-6"

client = anthropic.Anthropic(api_key=os.environ.get("ANTHROPIC_API_KEY"))
print("Client initialized:", client)
```

```python
# Basic message example
message = client.messages.create(
    model="claude-sonnet-4-6",
    max_tokens=1024,
    messages=[
        {"role": "user", "content": "Hello! Confirm the API is working in one sentence."}
    ]
)

print(message.content[0].text)
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)
```

```python
def chat(messages, system=None, temperature=1.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }
    
    if system:
        params["system"] = system
    
    message = client.messages.create(**params)
    return message.content[0].text
```

```python
messages = []

add_user_message(messages, "Generate five one sentence movie ideas")
response = chat(messages, temperature=0.5)
print("Response:", response)
answer = chat(messages, temperature=1)
print("Answer:", answer)

```

```python
messages = []

prompt = "Generate three different samples AWS CLI commands. Each should be very short"

add_user_message(messages, prompt)

text = chat(messages)

text.strip()
```
